# 📊 Notebook 01: Exploratory Data Analysis (EDA)
This notebook explores the IBM Telco Customer Churn dataset to understand features, distributions, correlation, and initial insights related to customer churn.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('../data/raw/telco_churn.csv')
df.head()
```

### 1. Check Class Imbalance
Let's see the proportion of churned versus retained customers.

```python
churn_counts = df['Churn'].value_counts(normalize=True)
print(churn_counts)

churn_counts.plot(kind='bar', color=['#2DD4BF', '#FF4B4B'])
plt.title('Churn Distribution (Target)')
plt.ylabel('Proportion')
plt.xlabel('Churned (Yes/No)')
plt.show()
```

### 2. Correlation Heatmap
We convert TotalCharges to numeric (handling spaces) and display correlation between numeric features.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(0, inplace=True)

sns.heatmap(df.select_dtypes('number').corr(), annot=True, cmap='coolwarm', fmt='.3f')
plt.title('Correlation Heatmap')
plt.show()
```

### 3. Churn Rate by Contract Type
Month-to-month contract types typically represent much higher churn risk than 1-year or 2-year contracts.

```python
contract_churn = df.groupby('Contract')['Churn'].apply(lambda x: (x == 'Yes').mean())
print(contract_churn)

contract_churn.plot(kind='bar', color='#6C63FF')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate')
plt.show()
```
